```python
# Function to transform images from Perspective View to Top-Dowm View(Bird'e Eye View)

def transform_prespective(image):
    src_points = []
    dst_points = []
    if image.shape[0] == 1550: 
        src_points = np.float32([
            [0, 1150], 
            [300, 1550],
            [2048, 1150], 
            [1800, 1550]
        ])
        dst_points = np.float32([
            [0, 0],
            [0, 1550], 
            [2048, 0],
            [2048, 1550]
        ])
    else:
        src_points = np.float32([
         [0, 1180],
         [0, 2048],
         [1550, 1180],
         [1550, 2048]
        ])
        dst_points = np.float32([
            [0, 0],
            [0, 2048],
            [1550, 0],
            [1550, 2048]
        ])
    M = cv2.getPerspectiveTransform(src_points, dst_points)
    if image.shape[0] == 1550: 
        Img = cv2.warpPerspective(image, M, (2048, 1550))
    else:
        Img = cv2.warpPerspective(image, M, (1550, 2048))
    
    return Img
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.subplot(1, 2, 1)
plt.imshow(cv2.imread(images[0]))
plt.title('Image vs BEV')


plt.subplot(1, 2, 2)
plt.imshow(transform_prespective(cv2.imread(images[0])))

plt.show()
```
